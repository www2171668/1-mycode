# 1、图框和图
——plt.figure()

    所有图像都必须位于figure对象中，figure可以有多个。用GUI打开方式时，同时出现多个窗口

    facecolor 底色
    figsize 设置图框大小，一般个位数就行
    num 图框编号，默认情况下从1开始排序，只有一个图框时不用写

```python
x = np.arange(-3,3,0.1)   # 间隔不够图形就不够顺滑
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()    # 1.1、创建figure()对象
plt.plot(x,y1)   # 1.2、创建plot图形①

plt.figure(num = 2,figsize = (6,1))   # 2.1、创建figure()对象，并设置编号2和大小(6,1)
plt.plot(x,y2)   # 2.2、创建plot图形②

plt.show()
```

##### 颜色、标记、线性
—— plt.plot()

    -》alpha 设置透明度
    -》linewidth / markersize 
    -》label 需要配合plt.legend()使用
    -》zorder 层数，层数越高，图像位置越高

```python
x = np.arange(-3,3,0.1)
y1 = np.sin(x)

plt.figure()
# plt.plot(x,y1,color = 'red')
# plt.plot(x,y1,color = 'red',marker = '>')   # -. 的形状
# plt.plot(x,y1,color = 'red',marker = '>',linestyle = '-.')   # markersize 标记点形状
# plt.plot(x,y1,color = 'red',marker = '>',linestyle = '-.',markersize = 10,alpha = 0.5)

#简写方式
plt.plot(x,y1,'ro--',markersize = 5,linewidth = 10, alpha = 0.5)  # r是颜色，o是标记（圆），--是线型
plt.show()
```

##### 刻度、标题、标签和图例
—— plt.legend() 显示plot的label

    loc 控制label位置

—— plt.xlabel() x轴标签 / plt.ylabel() y轴标签；★★ 

—— plt.title() 顶层标签 / plt.subtitle() 底层标签

—— plt.xlim() plt.ylim() 设置范围

—— plt.xticks(范围，刻度值) plt.yticks() 设置刻度

—— plt.gca() 获取图片信息，用于调用图片的属性

    .spines 边界属性，返回元祖列表，用['top']调用
        .set_color 设置边框颜色，

```python
x1 = [1,2,3]
y1 = [5,7,4]
x2 = [1,2,3]
y2 = [10,12,14]

plt.figure()
plt.plot(x1,y1,'ro-',label = '进口')   # label 右上角标签名；注意使用中文时，使用中文文本修正代码
plt.plot(x2,y2,'bo--',label = '出口')
plt.legend(loc='upper left')   # legend() 显示label

#设置标题标签
plt.xlabel('月份')     # x轴 标签
plt.ylabel('美元/亿')   # y轴 标签
plt.title('进出口数据')   # title  顶层标签

#设置范围
plt.xlim(0,6)
plt.ylim(0,15)

#设置刻度
plt.xticks(np.arange(1,6),[str(i)+'月' for i in range(1,6)])    #设置1-6的刻度；并对普通文字进行修改
plt.yticks(np.arange(2,16,2),['200','300','400','500','600','700','800'])    #设置2-16（间隔2）的刻度；并用列表数据去替换前面的8个刻度
                                                                        #间隔为2时，图中的间距也会加大
# 通常情况，就是用np.arange来控制刻度数量，然后用for循环 或者 列表 来重写刻度值

# 1、获取坐标轴信息
ax = plt.gca()

# 2、spines[] 设置边框 ，注意用[] 而不是（）
ax.spines['top'].set_color('none')    # set_color() 颜色设为none时，边框消失
ax.spines['right'].set_color('none')

plt.show()
```

# 2、subplot子图

##### 方法一
—— plt.subplot(行 列 位置) 创建子图框架，用于在figure对象下创建一个或多个subplot对象用于绘制plot图像   ★★★

    行 列 位置 都是从1开始计数

```python
x1 = [1,2,3]
y1 = [5,7,4]
x2 = [1,2,3]
y2 = [10,12,14]

plt.figure()
plt.subplot(221)   # 创建第一个子图，2行 2列 1位置
plt.plot(x1,y1,'ro--')

plt.subplot(224)   # 创建第二个子图
plt.plot(x2,y2,'bo-')

plt.show()
```

##### 方法二
—— fig.add_subplot(行 列 位置) 创建子图框架

—— plt.tight_layout() 自动调整子图参数，使之填充整个图像区域

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)   # 2行1列位置2

ax1.plot(np.random.randn(50).cumsum(),'g--')
ax2.plot(np.random.randn(50).cumsum(),'b--')
ax3.plot(np.random.randn(50).cumsum(),'r--')

plt.tight_layout()
plt.show()
```

—— .cumsum将矩阵的元素依次累加，返回由中间结果组成的数组

```python
print(np.arange(9).reshape(3,3))
print(np.arange(9).reshape(3,3).cumsum())
```

# 3、subplots 子图
—— plt.subplots(*)

    使用subplots(*) 时不用figure()来初始化图形
    返回两个变量①、fig作为figure对象
                ②、axes作为sub子图对象，在axes[ 索引号 ]上绘制plot图形

    -》sharex 多图共享x轴，sharey 多图共享y轴
    -》zorder 控制通道层，zorder越大，图像越在上层

```python
fig,axes = plt.subplots(4,1,sharex = True,sharey = True)   
# 也可以写成(nrows = 4,ncols = 1)，其中 nrows 行，ncols 列

#axes是一个列表
axes[0].plot(range(10),'ro-', ms=0.0001, zorder=4)   # ms控制标记大小 
axes[1].plot(range(10),'bo-',zorder=4)
axes[2].plot(range(10),'yo-')
axes[3].plot(range(10),'mo-')

plt.show()
```

—— .subplots_adjust() 调整plot子图形状

    wspace 子图间宽度间隔
    hspace 子图间高度间隔

```python
fig,axes = plt.subplots(2,3 ,sharex = True)

for i in range(2):        # 使用嵌套for循环绘制多子图的图形
    for j in range(3):
        axes[i][j].hist(np.random.randn(100),10)
        
plt.subplots_adjust(wspace =0.3,hspace = 0.3)
plt.show()
```
